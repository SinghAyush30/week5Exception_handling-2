Ans4) The LookupError class is a base class in Python's exception hierarchy that serves as a parent class for exceptions related to lookup operations. It is designed to handle errors that occur when attempting to access values from a collection, such as lists, dictionaries, or other data structures.

The LookupError class itself is not usually directly raised in code, but it provides a common interface and behavior for its derived subclasses, which include KeyError and IndexError.

```python
#i) Example for KeyError
try:
    my_dict = {"apple": "red", "banana": "yellow", "orange": "orange"}
    print(my_dict["apple"]) 
    print(my_dict["grape"])
except KeyError as e:
    print(e,"key comes under KeyError")
```

```python
#ii) Example for IndexError
try:
    my_list = [1, 2, 3, 4, 5]
    print(my_list[2])
    print(my_list[10]) 
except IndexError as e:
    print(e," comes under IndexError")
```

Ans5) ImportError is an exception that occurs when there is a problem in importing a module in Python. It is a common error that typically arises when the interpreter encounters an issue while trying to locate, load, or execute the specified module.
